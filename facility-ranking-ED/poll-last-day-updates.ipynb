```python
import pandas as pd
```

```python
from __future__ import print_function

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
```

```python
# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/drive.metadata.readonly']

creds = Credentials.from_authorized_user_file('token.json', SCOPES)
```

```python
service = build('drive', 'v3', credentials=creds)
```

```python
from datetime import datetime, timedelta
t = str((datetime.today() - timedelta(days=2)).isoformat())
# datetime.today().isoformat()
```

```python
files = []
page_token = None

while True:
    # pylint: disable=maybe-no-member
    response = service.files().list(q=f"modifiedTime > '{t}'",
                                    supportsAllDrives=True,
                                    includeItemsFromAllDrives=True,
                                    corpora='domain',
                     # fields='files'
                                    fields='files(owners/displayName, name, webViewLink, lastModifyingUser/displayName, modifiedTime)',
                                    pageToken=page_token
                                   ).execute()
    # for file in response.get('files', []):
    #     # Process change
    #     print(F'Found file: {file.get("name")}, {file.get("teamDriveId")}')
    files.extend(response.get('files', []))
    page_token = response.get('nextPageToken', None)
    if page_token is None:
        break
```

```python
df = pd.DataFrame.from_records(files)

df.modifiedTime = pd.to_datetime(df.modifiedTime)
# df = df[~df.name.str.contains('jpeg')].reset_index(drop=True)

df.sort_values('modifiedTime', ascending=False).head(30)
```
