```python
import pandas as pd
import difflib
```

```python
file_location = './data/Triage & ECG Training (Responses) - Form responses 1(1).csv'
```

```python
raw_df = pd.read_csv(file_location)
```

```python
raw_df
```

```python
course_type = 'ecg'
year_filter = pd.to_datetime(raw_df.Timestamp).dt.year.astype(str) == '2022'
course_filter = raw_df.Course.str.contains(course_type)

num_participants = raw_df.groupby(['County','Date','Course'])\
    .count()['Trainers Name'].reset_index().rename({'Trainers Name':'Num participants'}, axis=1)
num_participants
```

```python
num_participants['Date'] = pd.to_datetime(num_participants.Date, format="%d/%m/%Y")
```

# MEAL Report 2022

Looking at num triage

```python
year = '2022'
year_filt = num_participants.Date.dt.year.astype(str) == year
course_filt = num_participants.Course.str.lower().str.contains('triage')
course_filtered_df = num_participants[year_filt & course_filt]
print(course_filtered_df['Num participants'].sum())
course_filtered_df.sort_values('Num participants', ascending=False)
```

# Standardizing county names

```python
counties_file = '../kenya-counties/out.csv'
counties_df = pd.read_csv(counties_file, dtype=object)
counties_df.sample(3)
```

```python
num_participants.County = num_participants.County.apply(lambda x: difflib.get_close_matches(x, counties_df.County)[0])
```

```python
num_participants = num_participants.merge(counties_df[['County','Code']]).drop('County', axis=1)
```

```python
num_participants.to_csv('./data/output/triage_ecg.csv', index=False)
```
