```python
import pandas as pd
from difflib import get_close_matches

# Load the 'contacts_sheet' and 'unknowns' DataFrames
contacts_sheet = pd.read_csv("contacts_report.csv")
unknowns = pd.read_csv("unknowns.csv")

# Function to find closest matches and replace 'Title' with correct designation
def replace_unknown_titles(row, unknowns):
    first_name = row["First Name"]
    last_name = row["Last Name"]
    title = row["Title"]
    
    # Create a list of potential matches from 'unknowns' using 'get_close_matches'
    potential_matches = unknowns.apply(lambda x: (x["First Name"], x["Last Name"]), axis=1)
    
    # Find the closest match for 'First Name' and 'Last Name'
    closest_match = get_close_matches((first_name, last_name), potential_matches, n=1)
    
    if closest_match:
        matched_row = unknowns[(unknowns["First Name"] == closest_match[0][0]) & (unknowns["Last Name"] == closest_match[0][1])]
        if not matched_row.empty:
            matched_title = matched_row.iloc[0]["Title"]
            
            # Replace 'Title' with the matched title if 'Title' is 'Unknown'
            if title == "Unknown":
                return matched_title
    return title

# Apply the 'replace_unknown_titles' function to each row in 'contacts_sheet'
contacts_sheet["Corrected Title"] = contacts_sheet.apply(replace_unknown_titles, unknowns=unknowns, axis=1)

contacts_sheet["Title"] = contacts_sheet["Corrected Title"]
changes_occurred = contacts_sheet[contacts_sheet["Title"] != contacts_sheet["Corrected Title"]]
changes_occurred

# Drop the 'Corrected Title' column if you no longer need it
contacts_sheet.drop(columns=["Corrected Title"], inplace=True)
# Display the 'contacts_sheet' DataFrame with corrected titles
print(contacts_sheet)
```

```python
import pandas as pd
from difflib import get_close_matches

# Load the 'contacts_sheet' and 'unknowns' DataFrames
contacts_sheet = pd.read_csv("contacts_report.csv")
unknowns = pd.read_csv("unknowns.csv")

# Function to find the closest match for 'First Name' and 'Last Name'
def find_closest_match(row, unknowns):
    first_name = row["First Name"]
    last_name = row["Last Name"]
    
    # Create a list of potential matches from 'unknowns' using 'get_close_matches'
    potential_matches = unknowns.apply(lambda x: (x["First Name"], x["Last Name"]), axis=1)
    
    # Find the closest match for 'First Name' and 'Last Name'
    closest_match = get_close_matches((first_name, last_name), potential_matches, n=1)
    
    return closest_match[0] if closest_match else None

# Function to replace 'Title' with the matched title if 'Title' is 'Unknown'
def replace_unknown_title(row, unknowns):
    title = row["Title"]
    matched_row = find_closest_match(row, unknowns)
    
    if matched_row and title == "Unknown":
        return matched_row[2] if len(matched_row) > 2 else title  # Check if index 2 exists
    return title

# Apply the 'replace_unknown_title' function to each row in 'contacts_sheet'
contacts_sheet["Title"] = contacts_sheet.apply(replace_unknown_title, unknowns=unknowns, axis=1)

# Create a new DataFrame with changes (if any)
changes_occurred = contacts_sheet[contacts_sheet["Title"] != contacts_sheet["Title"]]

# Drop the 'Corrected Title' column if you no longer need it
contacts_sheet.drop(columns=["Corrected Title"], inplace=True)

# Display the 'contacts_sheet' DataFrame with corrected titles
print(contacts_sheet)

# Display the changes (if any)
print("Changes occurred:")
print(changes_occurred)
```
