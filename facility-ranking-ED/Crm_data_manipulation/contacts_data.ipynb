# View "unknowns" in the table

```python
import pandas as pd
contacts = pd.read_csv("contacts_report.csv")
```

```python
title_counts = contacts["Title"].value_counts()
for title, count in title_counts.items():
    print(f"{title}: {count}")
# Display the full range of counts for each unique title
title_counts
```

```python
contacts_filtered = contacts[["First Name", "Last Name", "Title"]].copy()
contacts_filtered["Title"].fillna("", inplace=True)
unknowns = contacts_filtered[contacts_filtered["Title"].str.contains("Unknown")]
unknowns.to_csv("unknowns.csv", index=False)
```

## Splitting tables and editing contact details

```python
import pandas as pd
main_cont = pd.read_csv('fif.csv')
main_cont.head(5000)
```

```python
title_counts = main_cont["Title"].value_counts()
for title, count in title_counts.items():
    print(f"{title}: {count}")
# Display the full range of counts for each unique title
title_counts
```

```python
# Assuming 'your_dataframe' is your DataFrame containing the data
# Take the first 5000 entries
first_5000 = main_cont.head(5000)

# Specify the file path and name where you want to save the CSV file
csv_file_path = 'first_5000_entries.csv'

# Convert the selected data to a CSV file
first_5000.to_csv(csv_file_path, index=False)

# Print a confirmation message
print(f"CSV file saved as '{csv_file_path}'")
```

```python
# Assuming 'main_cont' is your DataFrame
# Convert the "Mobile" column to string
main_cont['Mobile'] = main_cont['Mobile'].astype(str)

# Define a function to add "0" to the beginning of phone numbers
def add_leading_zero(Mobile):
    if Mobile.isdigit() and len(Mobile) == 9 and Mobile.startswith('7'):
        return '0' + Mobile
    return Mobile

# Apply the function to the "Mobile" column
main_cont['Mobile'] = main_cont['Mobile'].apply(add_leading_zero)
```

```python
# Apply a lambda function to check if a value can be converted to an integer
is_integer_mobile = first_5000["Mobile"].apply(lambda x: x.isdigit() if isinstance(x, str) else False)

# Check if all values in the column are integers
are_all_integers = is_integer_mobile.all()

# Print the result
print("All values in the 'Mobile' column are integers:", are_all_integers)
```

```python
# Filter the DataFrame to show where 'Mobile' column values are not integers
non_integer_mobile = first_5000[~is_integer_mobile]


non_integer_mobile["Mobile"]
```

```python
# Convert the "Mobile" column to integers, replacing non-integer values with NaN
first_5000["Mobile"] = pd.to_numeric(first_5000["Mobile"], errors="coerce")

# Check the data type of the "Mobile" column
print("Data type of 'Mobile' column after conversion:", first_5000["Mobile"].dtype)
```
